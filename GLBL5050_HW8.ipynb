```python

descriptions = data2['Description'].dropna() 


# Basic tokenization function
# Alternative tokenization function
from nltk.tokenize import word_tokenize

def basic_tokenize(text):
    return word_tokenize(text.lower())

# Tokenize the descriptions using the basic method
tokenized_descriptions = [basic_tokenize(desc) for desc in descriptions]

# Train the Word2Vec model
model = Word2Vec(tokenized_descriptions, vector_size=200, window=8, min_count=1, workers=4)

# economy_vec = "money" if "money" in model.wv else model.wv.index_to_key[0]
# economy_vec_similar = model.wv.most_similar(economy_vec)

# print(economy_vec_similar)

# Find and display words similar to 'economy'
if 'economy' in model.wv:
    print(f"Most similar words to 'economy':", model.wv.most_similar('economy'))
else:
    print("Word 'economy' not found in the vocabulary.")

# Find and display words similar to 'political'
if 'political' in model.wv:
    print(f"Most similar words to 'political':", model.wv.most_similar('political'))
else:
    print("Word 'political' not found in the vocabulary.")

# Find and display words similar to 'religious'
if 'religious' in model.wv:
    print(f"Most similar words to 'religious':", model.wv.most_similar('religious'))
else:
    print("Word 'religious' not found in the vocabulary.")




# # Example: Find words similar to 'attack'
# sample_word = 'attack' if 'attack' in model.wv else model.wv.index_to_key[0]  # Ensure sample word exists
# similar_words = model.wv.most_similar(sample_word)

# # Display similar words
# print(f"Most similar words to '{sample_word}':", similar_words)

# economy_vec = Word2Vec(sentences = description[) 

# print(economy_vec.wv.most_similar("economy", topn=8))

# similar_to_economy = model.wv.most_similar("kidnap", topn=8)
# print(similar_to_economy)


```

```python

economic_words = {"trade", "market", "businessman", "businesses", "business", "money", "economic", "finance", "economics", "marketplace", "financed", "wealthy", "businessmen", "economy", "markets", "wealth", "wealthiest", "marketing", "marketed", "economically", "trader", "moneymore", "finances", "businesswoman", "traders", "moneylender", "marketplaces", "trademark", "traded"}
religious_words = {"christian", "islamabad", "beliefs", "belief", "islamic", "muslim", "spiritual", "christians", "islam", "muslims", "islamique", "hindu", "islambouli", "faith", "religion", "faithful", "islamiya", "islamiyya", "islamist", "islamiyah", "christianity", "buddhist", "islami", "islamists", "islamia", "christiania", "hindus", "buddhists", "islamiah", "hindustan", "muslimeen", "muslimin", "islamica", "islamã", "islamluji", "hindugaon", "islammiyah", "islamically"}
political_words = {"states", "government", "power", "state", "powers", "stated", "powerful", "political", "governments", "statement", "politics", "statements", "politically", "powered", "policy", "governmental", "stateless", "authority", "diplomacy", "statehood", "powergel", "state4", "statemnt", "governmentã", "governmentno"}


perpetrator_categories = {
    "Al Qaeda": "religious",
    "ISIS": "religious",
    "Taliban": "religious",
    "Hamas": "religious",
    "FARC": "political",
    "IRA": "political",
    "Boko Haram": "religious",
    "ETA": "political",
    "Shining Path": "political",
    "Hezbollah": "religious, political"
}

# Add a new column for category labels
data2['Category'] = ''

def determine_categories(desc):
    categories = []
    words = nltk.word_tokenize(desc.lower())
    
    if any(word in economic_words for word in words):
        categories.append("economic")
    if any(word in religious_words for word in words):
        categories.append("religious")
    if any(word in political_words for word in words):
        categories.append("political")

    return ', '.join(categories) if categories else 'None'

# Function to assign categories based on perpetrator
def categorize_perpetrator(perpetrator):
    return perpetrator_categories.get(perpetrator, 'Unknown')

# Apply the function to the 'Description' column
data2['Category'] = data2['Description'].dropna().apply(determine_categories)

# Apply the function to the 'Perpetrator' column
data2['Category'] = data2['Perpetrator'].fillna('Unknown').apply(categorize_perpetrator)
data2
```
